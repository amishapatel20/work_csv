# Opening a csv file from an URL

```python
import pandas as pd
import requests
from io import StringIO

# Use raw URL
url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/master/IPL/matches.csv"

headers = {"User-Agent": "Mozilla/5.0"}
req = requests.get(url, headers=headers)

# Parse CSV
data = StringIO(req.text)
df = pd.read_csv(data)

# Show the data
df.head()
```
